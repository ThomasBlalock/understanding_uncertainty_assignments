In European roulette, there are 37 numbers, 0-36, on which one can bet combinations of 1-36 (zero always loses). The bet winnings are according to the number of slots/numbers included in the wager (e.g. betting on odds covers 18 slots). Lets call that K. The winnings are $36/K - 1$, so the total payout, including the original 1 unit bet is $36/K$. The chance of winning is $K/37$ and the chance of losing is $(37-K)/37$. So the expected value is $K/37 * 36/K - 1 = 37/36 - 1 = 0.0278$
